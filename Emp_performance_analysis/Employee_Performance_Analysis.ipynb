# IMPORTING LIBRARIES

```python
import pandas as pd
```

# LOADING DATA

```python
ep=pd.read_csv("Employee_Performance.csv")
```

```python
ep.head()
```

# CHECKING DATA FOR NULL VALUE

```python
ep.isnull().sum()
```

# TOP 10 PRODUCTIVE EMPLOYEE-BASED ON PERFORMANCE

```python
PRODUCTIVITY_PORFOR = ep['Performance_Score'] / ep['Monthly_Salary'] * 10000
PRODUCTIVITY_PORFOR = pd.DataFrame({
    'Employee_ID': ep['Employee_ID'],
    'Productivity': PRODUCTIVITY_PORFOR
})
PRODUCTIVITY_PORFOR = PRODUCTIVITY_PORFOR.sort_values(by='Productivity', ascending=False)
PRODUCTIVITY_PORFOR.head(10)
```

# 10 LEAST PRODUCTIVE EMPLOYEE BASED ON PERFORMANCE

```python
PRODUCTIVITY_PORFOR = PRODUCTIVITY_PORFOR.sort_values(by='Productivity', ascending=True)
PRODUCTIVITY_PORFOR.head(10)
```

# TOP 10 PRODUCTIV EMPLOYEE PRODUCTIVITY BASE ON EXPERIENCE

```python
PRODUCTIVITY_EXP=ep['Performance_Score']/ep['Experience (Years)']
PRODUCTIVITY_EXP = pd.DataFrame({
    'Employee_ID': ep['Employee_ID'],
    'Productivity_Exp': PRODUCTIVITY_EXP
})
PRODUCTIVITY_EXP.sort_values(by='Productivity_Exp',ascending=False).head(10)
```

# 10 LEAST PRODUCTIVE EMPLOYEE BASED ON EXPERIENCE

```python
PRODUCTIVITY_EXP.sort_values(by='Productivity_Exp',ascending=True).head(10)
```

# TOP 10 PRODUCTIVE EMPLOYEE BASED ON SALARY

```python
Productivity_Salary = ep[['Employee_ID', 'Performance_Score', 'Monthly_Salary']].copy()
Productivity_Salary['Productivity_on_salary'] = (
    Productivity_Salary['Performance_Score'] /
    Productivity_Salary['Monthly_Salary']
) * 1000
Productivity_Salary = Productivity_Salary.sort_values(
    by='Productivity_on_salary',
    ascending=False
)
Productivity_Salary.drop(columns=['Performance_Score','Monthly_Salary']).head(10)
```

# LEAST 10 PRODUCTIVE EMPLOYEE ON SALARY BASED

```python
Productivity_Salary = Productivity_Salary.sort_values(
    by='Productivity_on_salary',
    ascending=True
)
Productivity_Salary.drop(columns=['Performance_Score','Monthly_Salary']).head(10)
```

# TOP 10 PRODUCTIVE EMPLOYEES ON DEPARTMENT BASED

```python
Department_productivity = pd.DataFrame({
    'Employee_ID': ep['Employee_ID'],
    'Department': ep['Department'],
    'Monthly_Salary': ep['Monthly_Salary'],
    'Productivity_on_salary': Productivity_Salary['Productivity_on_salary']
})
Department_productivity.sort_values(by='Productivity_on_salary',ascending=False).head(10)
```

# LAEST 10 PRODUCTIVE EMPLOYEES ON DEPARTMENT BASED

```python
Department_productivity.sort_values(by='Productivity_on_salary',ascending=True).head(10)
```
