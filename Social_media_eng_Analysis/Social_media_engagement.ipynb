# importing Libraries

```python
import pandas as pd
```

```python
fb=pd.read_csv(r"Social-media-dataset-samples-main\Facebook-datasets.csv")
ig=pd.read_csv(r"Social-media-dataset-samples-main\Instagram-datasets.csv")
tt=pd.read_csv(r"Social-media-dataset-samples-main\TikTok-datasets.csv")
tw=pd.read_csv(r"Social-media-dataset-samples-main\Twitter- datasets.csv")
```

# ADDING A PLATFORM COLUMN IN DATASET

```python
fb['Platform'] = 'Facebook'
ig['Platform'] = 'Instagram'
tt['Platform'] = 'TikTok'
tw['Platform'] = 'Twitter'
```

# COMBINING ALL DATASET IN ONE:

```python
df = pd.concat([fb, ig, tt, tw], ignore_index=True)
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns = ['url', 'post_url', 'user_url', 'comment_user_url', 'commenter_url', 'comment_url','external_url', 'external_image_urls', 'profile_image_link', 'parent_post_details',
    'attached_files', 'source_type', 'biography', 'name', 'id', 'videos', 'photos', 'hashtags', 'tagged_users', 'quotes', 'bookmarks', 'description'])
df.head()
```

```python
df.info()
```

# CALCULATING ENGAGEMENT

## COUNTING REPLIES AND COMMENTS

```python
df['comment_count'] = df['comment'].notna().astype(int)
df['reply_count'] = df['replies'].notna().astype(int)
```

# CALCULATING TOTAL ENGAGEMENT

```python
df['Total_Engagement'] = (
    df[['likes', 'reposts', 'views',
        'num_likes', 'likes_number',
        'num_replies', 'replies_number']]
    .fillna(0)
    .sum(axis=1)
    + df['comment_count']
    + df['reply_count']
)
df.groupby('Platform')['Total_Engagement'].sum().sort_values(ascending=False)
```

# POST PASTING PATTERN BY PLATFORM

```python
df.groupby('Platform').agg(
    total_posts=('post_id', 'count'),
    avg_engagement=('Total_Engagement', 'mean')
).sort_index(ascending=False)
```
