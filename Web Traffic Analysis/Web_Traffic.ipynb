# IMPORTING REQURIED LIBRARIES

```python
import pandas as pd
```

# IMPORTING DATA SET

```python
web=pd.read_csv("website_wata.csv")
web.head()
```

# OVERVIEW OF DATASET

```python
web.info()
```

# CHECKING FOR ANY NULL VALUES IN DATASET

```python
web.isnull().values.any()
```

# FINDING THE AVERAGES, MIN AND MAX VALUES 

```python
web.describe().drop(['std','count','25%','50%','75%'])
```

# FINDING THE TRAFFIC SOURCE NUMBERS

```python
web['Traffic Source'].value_counts()
```

# ANALYSING THE PAGE VIEWS, FINDING 10 MOST VIEWS 

```python
web[['Page Views']].sort_values(by='Page Views',ascending=False).head(10)
```

# VISUALIZING TRAFFIC SOURCES IN PERCENTAGES

```python
import matplotlib.pyplot as plt
```

```python
web['Traffic Source'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Traffic Source Distribution")
plt.ylabel('')
plt.show()
```

# FINDING TOP PAGES WITH BOUNCE RATE

```python
high_bounce = web[web['Bounce Rate'] > 0.5]
high_bounce[['Page Views','Bounce Rate']].sort_values(by='Bounce Rate',ascending=False).head(10)
```

# FINDING SESSION DURATION

```python
import seaborn as sns
sns.scatterplot(x='Page Views', y='Session Duration', data=web)
plt.title("Session Duration vs Page Views")
plt.show()
```

# FINDING CONVERSION RATE

```python
low_conversion = web[web['Conversion Rate'] < 0.5]
print(low_conversion[['Page Views','Conversion Rate']])
```
